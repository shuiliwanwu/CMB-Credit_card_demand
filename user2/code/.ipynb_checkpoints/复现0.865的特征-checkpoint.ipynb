```python
import pandas as pd
train_log = pd.read_csv('../train/train_log.csv',sep='\t')
test_log = pd.read_csv('../test/test_log.csv',sep='\t')
```

```python
train_log.head()
```

```python
log = pd.concat([train_log,test_log])
```

```python
log['time2'] = log.OCC_TIM.apply(lambda x: pd.to_datetime(x))
log['day'] = log.time2.apply(lambda x: x.day)
```

```python
log['count'] =1
t1 = log.groupby('EVT_LBL')['count'].agg({'evt_sum':'sum'}).reset_index()
```

```python
t3 = log.groupby(['USRID','EVT_LBL'])['count'].agg({'user_this_evt_count':sum}).reset_index()
t3.head()
```

```python
t2['evt_name'] =t2['EVT_LBL']
```

```python
t2.head()
```

```python
t4  = pd.merge(t3,t2,how='left')
```

```python
t4['evt_name'].fillna('others',inplace=True)
```

```python
t4.head()
```

```python
t5 = t4.groupby(['USRID','evt_name',])['user_this_evt_count'].sum().reset_index()
```

```python
t5.head()
```

```python
tmp = pd.pivot_table(t5,index=['USRID'],columns=['evt_name'],values=['user_this_evt_count'],fill_value=0,margins=False)
s= tmp['user_this_evt_count']
s.columns = s.columns.map(lambda x:'evt_'+str(x))
s.columns.name=None
s = s.reset_index()
feature_all_count = t5.groupby('USRID')['user_this_evt_count'].sum().reset_index()
```

```python
s.head() #s代表  每个用户在点击数大于100的下的 点击数
```

```python
s.to_csv('final_feature.csv',index=None,sep='\t')
```

```python
log.head()
```

```python
log['time2'] = log.OCC_TIM.apply(lambda x: pd.to_datetime(x))
log['day'] = log.time2.apply(lambda x: x.day)#第多少天
log['dayofweek']  = log.time2.apply(lambda x: x.isoweekday()) #周几
log['isWen'] = log.dayofweek.apply(lambda x: 1 if x==3 else 0) #是否是周三
log['isweekend'] = log.dayofweek.apply(lambda x: 1 if x in [6, 7] else 0)#是否是周末
log['exactTime'] = log.OCC_TIM.apply(lambda x: x[-8:])

log['promotiontime1'] = log.exactTime.apply(lambda x: 1 if '09:30:00'<=x<='10:30:00' else 0) #周三10点
log['promotiontime2'] = log.exactTime.apply(lambda x: 1 if '14:50:00'<=x<='15:10:00' else 0)#下午三点
```

```python
log.head()
```

```python
log['promotiontime1'] = log['promotiontime1']&log['isWen']  #周三10点的活动时间
#log['promotiontime2'] #每天下午三点
```

```python
# t1 = log.groupby(['USRID'],as_index=False)['isweekend'].agg({'weekend':sum}) #周末 点击次数
t2 = log.groupby(['USRID'],as_index=False)['promotiontime1'].agg({'promotiontime1':sum}) #活动时间1 周三10点 次数
t3 = log.groupby(['USRID'],as_index=False)['promotiontime2'].agg({'promotiontime2':sum}) #活动时间2  每天下午3点   次数
# t4 = log.groupby(['USRID'],as_index=False)['isWen'].agg({'Wen':sum}) #周三 点击次数
# t5 = log.groupby(['USRID'],as_index=False)['EVT_LBL'].agg({'EVT_LBL_len':len}) #某人总的 点击次数
# t6 = log.groupby(['USRID'],as_index=False)['EVT_LBL'].agg({'EVT_LBL_set_len':lambda x:len(set(x))}) #某人不重复的点击次数
```

```python
feature = pd.merge(t2,t3,on='USRID')
# feature = pd.merge(feature,t3,on='USRID')
# feature = pd.merge(feature,t4,on='USRID')
# feature = pd.merge(feature,t5,on='USRID')
# feature = pd.merge(feature,t6,on='USRID')
```
